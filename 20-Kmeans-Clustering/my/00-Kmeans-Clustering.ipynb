```python
X=pd.get_dummies(df,dtype=int)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
scaled_X=scaler.fit_transform(X)
```

```python
scaled_X
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=2,random_state=101)
```

```python
labels=model.fit_predict(scaled_X)
```

```python
labels
```

```python
X["Cluster"]=labels
```

```python
X
```

```python
X.corr()
```

```python
plt.figure(figsize=(12,7))
X.corr()["Cluster"][:-1].sort_values().plot(kind="bar")
```
