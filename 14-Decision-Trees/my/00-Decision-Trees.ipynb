```python
X=pd.get_dummies(df.drop('species',axis=1),dtype=int,drop_first=True)
```

```python
y=df["species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(y_test,base_pred)
```

```python
print(classification_report(y_test,base_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test,base_pred),display_labels=['Adelie', 'Chinstrap', 'Gentoo']).plot()
```

```python
pd.DataFrame(data=model.feature_importances_,index=X.columns,columns=["Feature Importance"]).sort_values("Feature Importance")
```

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi=150)
    plot_tree(model,filled=True,feature_names=X.columns);
```

```python
report_model(model)
```

```python
pruned_tree=DecisionTreeClassifier(max_depth=2)
```

```python
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

```python
max_leaf_tree=DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
max_leaf_tree.fit(X_train,y_train)
```

```python
report_model(max_leaf_tree)
```

```python
entropy_tree=DecisionTreeClassifier(criterion='entropy')
```

```python
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
