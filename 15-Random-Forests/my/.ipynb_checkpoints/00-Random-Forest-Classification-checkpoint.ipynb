```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../DATA/penguins_size.csv")
```

```python
df
```

```python
df = df.dropna()
df.head()
```

```python
df.isnull().sum()
```

```python
X=pd.get_dummies(df.drop("species",axis=1),drop_first=True,dtype=int)
```

```python
X
```

```python
y=df["species"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=10,max_features='sqrt',random_state=101)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_test)
```

```python
print(classification_report(y_test,y_test))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test,y_test),display_labels=rfc.classes_).plot()
```
