```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./data-files/dm_office_sales.csv")
```

```python
df
```

```python
sns.rugplot(x="salary",data=df)
```

```python
sns.rugplot(x="salary",data=df,height=0.5)
```

```python
sns.displot(data=df,x="salary",kde=True)
```

```python
sns.histplot(data=df["salary"])
```

```python
# sns.set_style("ticks")
sns.displot(data=df,x="salary",bins=20,color="red",edgecolor="blue",ls="--",rug=True)
```

```python
sns.kdeplot(data=df,x="salary")
```
