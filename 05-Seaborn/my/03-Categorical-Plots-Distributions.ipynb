```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./data-files/StudentsPerformance.csv")
```

```python
df
```

```python
plt.figure(figsize=(4,2),dpi=100)
sns.boxplot(x="math score",data=df)
```

```python
plt.figure(figsize=(10,6),dpi=100)
sns.boxplot(x="math score",data=df,y="parental level of education")
```

```python
plt.figure(figsize=(10,6),dpi=100)
sns.boxplot(y="math score",data=df,x="parental level of education")
```

```python
plt.figure(figsize=(10,6),dpi=100)
sns.boxplot(x="math score",data=df,y="parental level of education",hue="gender",orient="h",width=0.5)
plt.legend(bbox_to_anchor=(1.05,1))
```
