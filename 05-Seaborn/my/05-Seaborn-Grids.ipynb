```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./data-files/StudentsPerformance.csv")
```

```python
df
```

```python
sns.catplot(data=df,x="gender",y="math score",kind="box")
```

```python
sns.catplot(data=df,x="gender",y="math score",kind="box",row="lunch")
```

```python
sns.catplot(data=df,x="gender",y="math score",kind="box",row="lunch",col="test preparation course")
```

```python
g=sns.PairGrid(df)
g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.kdeplot,lw=2)
g=g.map_lower(sns.kdeplot,color="red")
```

```python
g=sns.PairGrid(df,hue="gender")
g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.kdeplot,lw=2)
g=g.map_lower(sns.kdeplot,color="red")
```

```python
g=sns.FacetGrid(data=df,row="gender",col="lunch")
g= g.map(plt.scatter,"math score","reading score",edgecolor="w")
```
