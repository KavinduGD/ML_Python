```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./data-files/dm_office_sales.csv")
```

```python
df
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.countplot(data=df,x="division")
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.countplot(data=df,x="level of education")
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.countplot(data=df,x="level of education",hue="division")
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.countplot(data=df,x="level of education",hue="division",palette="Set2")
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.barplot(data=df,x="division",y="salary",estimator=np.mean,errorbar="sd",hue="level of education")
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.barplot(data=df,x="division",y="salary",estimator=np.mean,errorbar="sd",hue="level of education")
plt.legend(bbox_to_anchor=(1.05,1))
```
