```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])

axes1.plot(a,b)


axes1.set_xlabel("x lable")
axes1.set_ylabel("y lable")
axes1.set_title("Ghraph")

axes2=fig.add_axes([1,-0.3,0.3,0.3])
axes2.plot(a,b)

axes2.set_xlabel("x lable")
axes2.set_ylabel("y lable")
axes2.set_title("Ghraph")
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)

```
