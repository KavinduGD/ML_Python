```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
a=np.linspace(0,10,11)
b=a**4
```

```python
x=np.arange(0,10)
y=x*2
```

```python
x
```

```python
y
```

```python
fig,axes=plt.subplots()
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes=plt.subplots(2,2,figsize=(5,5),dpi=100)
axes[0,0].plot(x,y)
axes[0,0].set_xlabel("x labe")
axes[0,1].plot(a,b)
axes[1,0].plot(a,b)
axes[1,1].plot(x,y)

fig.suptitle("This is super title")

fig.subplots_adjust(wspace=1,hspace=0.6)

fig.savefig("./sub.jpg",bbox_inches="tight")
```
