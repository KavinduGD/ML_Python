```python
import matplotlib.pyplot as pt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,11)
```

```python
x
```

```python
fig=pt.figure()
axes= fig.add_axes([0,0,1,1])

axes.plot(x,x,label="X vs X")
axes.plot(x,x**2,label="X vs X**2")

axes.legend(loc=(1.05,0.9))
```

```python
fig=pt.figure(figsize=(10,7))
axes= fig.add_axes([0,0,1,1])

lines=axes.plot(x,x,label="X vs X",lw=8,marker="s",ms=20,markerfacecolor="red",markeredgewidth=10,markeredgecolor="yellow")


```
